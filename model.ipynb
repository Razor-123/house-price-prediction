```python
train_df22 = train_df21.drop(['locality','activation_date','amenities','lease_type','furnishing','parking','facing','water_supply','building_type'],axis=1)
train_df22
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train = scaler.fit_transform(train_df22)
multiplied_by = scaler.scale_[1983]
added = scaler.min_[1983]
multiplied_by
```

```python
scaled_train.shape
```

```python
scaled_train_df = pd.DataFrame(scaled_train, columns=train_df22.columns.values)
```

```python
scaled_train_df
```

```python
model = Sequential()

model.add(Dense(50, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
X = scaled_train_df.drop('rent', axis=1).values
Y = scaled_train_df[['rent']].values

# Train the model
model.fit(
    X[10:],
    Y[10:],
    epochs=10,
    shuffle=True,
    verbose=2
)
```

```python
prediction = model.predict(X[:1])
y_0 = prediction[0][0]
print('Prediction with scaling - {}',format(y_0))
y_0 -= added
y_0 /= multiplied_by
print("Housing Price Prediction  - ${}".format(y_0))
```
